## SciPy

![SciPy](bilder/scipy.png)

* Scientific Python
* Är en organisation och ett helt ekosystem (där NumPy, Pandas, MatplotLib ingår)
* SciPy core:
    * Scientific units and constants
    * Integration (lösare för integraler och differentialekvationer)
    * Clustering algoritmer
    * Optimization - root finders mm
    * Interpolation
    * Fast Fourier Transforms
    * Signal Processing
    * __Linear Algebra (vektorer, matriser, funktioner och lösare)__
    * Sparse eigenvalue problems with ARPACK
    * Compressed Sparse Graph Routines 
    * Spatial data structures and algorithms
    * __Statistics (scipy.stats)__
    * n-dimensionell bild processning 
    * Fil IO (MATLAB mfl "vetenskapliga" filformat)
